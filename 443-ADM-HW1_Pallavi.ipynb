### Exercise 1 : Online Advertising Sensitivity Analyses

#### Decision variables
Let $i={1,2,3,4}$ be the index for the advertisers such that
* $i=1$ is GEICO
* $i=2$ is Delta
* $i=3$ is T-Mobile
* $i=4$ is Capital One

Let $j={1,2}$ be the index for the category, such as $j=1$ is the Sports categorty and $j=2$ is the National category.

So $x_{11}$ is the number of impressions for Geico in Sports and $x_{12}$ is the number of impressions for GEICO in the National category

#### Objective function
The objective is to maximize the advertising revenues. These revenues are driven by the cost per click charged to the advertisers, i.e., \$2.4 and the click-through rates.

Let $\kappa_{ij}$ be the CTR of advertiser $i$ in category $j$. For instance, the CTR of T-Mobile in the sports category is $\kappa_{31}=1.0\%$, whereas its CTR in the National category is $\kappa_{32}=3.0\%$.

The objective function is thus:
\begin{equation}
Objective=2.4\times \sum_{i=1}^4\sum_{j=1}^{2}  \kappa_{ij}\times x_{ij}
\end{equation}

#### Constraints
The first table provides the following constraints:
\begin{align}
x_{11}&\geq 2,000,000\\
x_{12}&\geq 1,000,000\\
x_{22}&\geq 1,000,000\\
x_{21}+x_{22}&\geq 2,000,000\\
x_{31}&\geq 1,000,000\\
x_{32}&\geq 1,000,000\\
x_{31}+x_{32}&\geq 3,000,000\\
x_{41}+x_{42}&\geq 2,000,000\\
\end{align}

"Assume that the Sports section gets six million views per day and the National section
get five million views per day"
This statement gives two constraints:
\begin{align}
\sum_{i=1}^{4}x_{i1}&\leq 6,000,000\\
\sum_{i=1}^{4}x_{i2}&\leq 5,000,000\\
\end{align}

Non-negativity constraints are:
\begin{equation}
x_{ij}\geq 0
\end{equation}

#### Optimal Solution

```python
#Uncomment the below 2 pip commands when running on google colab
#!pip install cvxpy
#!pip install cvxopt
#Load required packages
import numpy 
import math
import cvxpy as cvx
from numpy import *
from cvxpy import *
import pandas as pd

#Define variables
kappa=matrix([[2.5,2.0,1.0,1.5],[0.8,1.0,3.0,2.0]])
kappa1=kappa[0]
kappa2=kappa[1]
x1=cvx.Int(4,1)
x2=cvx.Int(4,1)

# Objective Function
Z1=sum_entries(kappa1*x1)*2.4/100
Z2=sum_entries(kappa2*x2)*2.4/100
objective = cvx.Maximize((Z1+Z2))

# Constraints
c1=(sum_entries(x1))<=6*1000000 # Capacity on Sports
c2=(sum_entries(x2))<=5*1000000 # Capacity on Sports
c3=x1[0]>=2*1000000 # Geico must have at least 2m in sports
c4=x2[0]>=1*1000000 # Geico must have at least 1m in sports
c5=x2[1]>=1*1000000 # Delta must have at least 1m in sports
c6=x1[1]+x2[1]==2*1000000 # Delta's total # of impressions must be equal to 2m total
c7=x1[2]>=1*1000000# Tmobile must have at least 1m impression in sports
c8=x2[2]>=1*1000000# Tmobile must have at least 1m impression in national
c9=x1[2]+x2[2]==3*1000000 # Tmobile's total # of impressions must be equal 3m total
c10=x1[3]+x2[3]==2*1000000# Capital One's impression equal 2m
c=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,x1>=0,x2>=0]

#Solution
prob = cvx.Problem(objective, c)
result = prob.solve()
print('The optimal revenue is ')
print(prob.value)
print('The optimal allocation in the sports category is')
print(x1.value)
print('The optimal allocation in the national category is')
print(x2.value)
```

```python
#Define variables
x1=cvx.Int(4,1)
x2=cvx.Int(4,1)

#Constraints
c1=(sum_entries(x1))<=6*1000000#Capacity on Sports
c2=(sum_entries(x2))<=5*1000000#Capacity on Sports
c3=x1[0]>=2*1000000 # Geico must have at least 2m in sports
c4=x2[0]>=1*1000000 # Geico must have at least 1m in sports
c5=x2[1]>=1*1000000 # Delta must have at least 1m in sports
c6=x1[1]+x2[1]==2*1000000 # Delta's total # of impressions must be equal to 2m total
c7=x1[2]>=1*1000000# Tmobile must have at least 1m impression in sports
c8=x2[2]>=1*1000000# Tmobile must have at least 1m impression in national
c9=x1[2]+x2[2]==3*1000000 # Tmobile's total # of impressions must be equal 3m total
c10=x1[3]+x2[3]==2*1000000# Capital One's impression equal 2m
c=[c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,x1>=0,x2>=0]

#Change click-thru rates here to run sensitvity analysis
arr_i = numpy.arange(2.0, 3.1, 0.1)
geico_s = pd.DataFrame()
for i in range(0, len(arr_i)):
    kappa=matrix([[arr_i[i],2.0,1.0,1.5],[0.8,1.0,3.0,2.0]])
    kappa1=kappa[0]
    kappa2=kappa[1]
        
    #Objective Function
    Z1=sum_entries(kappa1*x1)*2.4/100
    Z2=sum_entries(kappa2*x2)*2.4/100
    objective = cvx.Maximize((Z1+Z2))
        
    #Solution
    prob = cvx.Problem(objective, c)
    result = prob.solve()
    print('######################## GEICO - Sports CTR as %s%% #######################'%arr_i[i])
    print('The optimal revenue is')
    print(prob.value)
    print('The optimal allocation in the sports category is')
    print(x1.value)
    print('The optimal allocation in the national category is')
    print(x2.value)
        
    #Appending to matrix
    geico_s = geico_s.append([[arr_i[i],prob.value]]) 
```

```python
#To view the results in a table - sorted the table by top to bottom Revenue 
geico_s.columns=['Sports CTR','Optimal Revenue']
geico_s.sort_values(by=['Optimal Revenue'],ascending=False)
```

#### For separate optimization on Plant A and B
##### Decision variables
Let $i={1,2}$ be the index for the plants such that

Let $j={1,2}$ be the index for the product, such as $j=1$ is the standard product and $j=2$ is the deluxe product.

So $x_{11}$ is the units of standard product from Plant A and $x_{12}$ is the units of deluxe product from Plant A.

##### Objective function
The 1st objective of the firm is to maximize the profit from each plant based on the total units of both the products produced per week. Standard product has a profit contribution of \$10 and deluxe product of \$15.

For Plant A, the 1st objective function is:
\begin{equation}
Profit=10\times x_{11}+15\times x_{12}
\end{equation}

Similarly for Plant B, the 2nd objective function is:
\begin{equation}
Profit=10\times x_{21}+15\times x_{22}
\end{equation}

##### Constraints
For Plant A:-

Based on the processing times (grinding and polishing respectively):
\begin{align}
4\times x_{11}+2\times x_{12}&\leq 80\\
2\times x_{11}+5\times x_{12}&\leq 60\\
\end{align}

Based on raw material (75 kgs cap per week):
\begin{align}
4\times x_{11}+4\times x_{12}&\leq 75\\
\end{align}

Non-negativity constraints are:
\begin{equation}
x_{1j}\geq 0
\end{equation}

For Plant B:-

Based on the processing times (grinding and polishing respectively):
\begin{align}
5\times x_{21}+3\times x_{22}&\leq 60\\
5\times x_{21}+6\times x_{22}&\leq 75\\
\end{align}

Based on raw material (45 kgs cap per week):
\begin{align}
4\times x_{21}+4\times x_{22}&\leq 45\\
\end{align}

Non-negativity constraints are:
\begin{equation}
x_{2j}\geq 0
\end{equation}

##### Solution for Plant A

```python
#Define Variables
x11 = cvx.Int()# Standard Product
x12 = cvx.Int()# Deluxe Product

#Constraints
c1= 4*x11+2*x12<= 80 
c2= 2*x11+5*x12<= 60 
c3= 4*x11+4*x12<= 75 
c4= x11>=0 
c5= x12>=0
con=[c1,c2,c3,c4,c5]

#Objective
OF=10*x11 + 15*x12
objective = cvx.Maximize(OF)

#Solution
prob = cvx.Problem(objective, con)
result = prob.solve()

print('optimal profit')
print(prob.value)

print('Standard production level')
print (x11.value)

print('Deluxe production level')
print (x12.value)
```

#### Solution for Joint optimization

##### Decision variables
Let $i={1,2}$ be the index for the plants such that

Let $j={1,2}$ be the index for the product, such as $j=1$ is the standard product and $j=2$ is the deluxe product.

So $x_{11}$ is the units of standard product from Plant A and $x_{12}$ is the units of deluxe product from Plant A.

##### Objective function
The objective of the firm is to maximize the profit together from both plants based on the total units of both the products produced per week. Standard product has a profit contribution of \$10 and deluxe product of \$15.

The objective function is:
\begin{equation}
Profit=10\times \sum_{i=1}^2 x_{i1}+ 15\times \sum_{i=1}^{2} x_{i2}
\end{equation}

##### Constraints
Based on the processing times (grinding and polishing respectively):
\begin{align}
4\times x_{11}+2\times x_{12}&\leq 80\\
2\times x_{11}+5\times x_{12}&\leq 60\\
5\times x_{21}+3\times x_{22}&\leq 60\\
5\times x_{21}+6\times x_{22}&\leq 75\\
\end{align}

Based on raw material (Ttotal 120 kgs cap per week):
\begin{align}
4\times \sum_{i=1}^2\sum_{j=1}^{2} x_{ij}&\leq 120\\
\end{align}

Non-negativity constraints are:
\begin{equation}
x_{ij}\geq 0
\end{equation}

```python
#Based on joint constraint on Raw material
#Total 120 can be distributed based on the final optimal units produced by each plant
#Define Variables
x11 = cvx.Int()# Plant A - Standard Product
x12 = cvx.Int()# Plant A - Deluxe Product
x21 = cvx.Int()# Plant B - Standard Product
x22 = cvx.Int()# Plant B - Deluxe Product

#Constraints
c1= 4*x11+2*x12<= 80 
c2= 2*x11+5*x12<= 60 
c3= 5*x21+3*x22<= 60 
c4= 5*x21+6*x22<= 75 
c5= 4*x11+4*x12+4*x21+4*x22<= 120 
c6= x11>=0 
c7= x12>=0
c8= x21>=0 
c9= x22>=0
con=[c1,c2,c3,c4,c5,c6,c7,c8,c9]

#Objective
OF=10*x11 + 15*x12 + 10*x21 + 15*x22
objective = cvx.Maximize(OF)

#Solution
prob = cvx.Problem(objective, con)
result = prob.solve()

print('optimal profit')
print(prob.value)

print('Plant A - Standard production level')
print (x11.value)

print('Plant A - Deluxe production level')
print (x12.value)

print('Plant B - Standard production level')
print (x21.value)

print('Plant B - Deluxe production level')
print (x22.value)
```

```python
# Assign ranks based on profitability (highest-1, 2nd highest-2,..so on) for greedy approach
hotel['Profit_Rank'] = hotel['Profitability'].rank(ascending=False)
hotel['Profit_Rank'] = hotel['Profit_Rank'].astype('int')
hotel
```

```python
# Define the Greedy algo function to add the list of most to least profitable hotels until we exhaust the budget
def greedy_app(budget):
    hotel_list = pd.DataFrame()
    for i in range(1,16):
        price = hotel.loc[hotel['Profit_Rank']==i,'Price'].iloc[0]
        hid = hotel.loc[hotel['Profit_Rank']==i,'Hotel'].iloc[0]
        location = hotel.loc[hotel['Profit_Rank']==i,'Location'].iloc[0]
        profit = hotel.loc[hotel['Profit_Rank']==i,'Profitability'].iloc[0]
        
        if (budget >= price):
            hotel_list = hotel_list.append([[hid, location, price, profit, i]]) 
            budget = budget - price
    
    hotel_list.columns = ['Hotel','Location','Price','Profitability','Profit_Rank']
    return hotel_list
```

So, if there is a constraint then maybe the next set of hotels could be from LA which will give lower profit. There are some hotels in other cities like Fresno and Long Beach where the profitability is very high but since the price is also very high, it's less likely for the optimization model to choose them so that the budget is not exhausted early (like in the greedy approach).

Thus, the additional constraints on the cities of Fresno, LA and South Lake Tahoe are:

\begin{align}
y_{1}+y_{2}+y_{3}+y_{4}&\leq 2\\
y_{6}+y_{7}+y_{8}&\leq 2\\
y_{9}+y_{10}+y_{11}+y_{12}+y_{13}+y_{14}+y_{15}&\leq 2\\
\end{align}

```python
#Define Variables
n = 16 # Number of hotels
y = cvx.Bool(n)

# from the data
x = hotel.iloc[:,2].values
z = hotel.iloc[:,7].values

#Constraints
c1= sum_entries(x*y)<= 10000000
c2= sum_entries(y)<= n
c3= sum_entries(y[1:5])<= 2
c4= sum_entries(y[6:9])<= 2
c5= sum_entries(y[9:])<= 2
con=[c1,c2,c3,c4,c5]

#Objective
OF=sum_entries(z*y)
objective = cvx.Maximize(OF)

#Solution
prob = cvx.Problem(objective, con)
result = prob.solve(solver=GLPK_MI)

print('Optimal profit')
print(prob.value)

print('Decision to buy each hotel')
print (y.value)
```
